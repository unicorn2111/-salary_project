```python
import pandas as pd
df = pd.read_csv('glassdoor_jobs.csv') 
```

```python
#SALARY PARSING
#  new column shows 1 if salary is provided hourly else 0
df['hourly'] = df['Salary Estimate'].apply(lambda x:1 if 'per hour' in x.lower() else 0)

# new column shows 1 if salary is employer provided else 0
df['employer provided'] = df['Salary Estimate'].apply(lambda x:1 if 'Employer provided salary:' in x.lower() else 0)

# seizing out all rows that had -1 in salary estimate column 
df = df[df['Salary Estimate']!='-1']

# taking out redundant text after the salary estimate
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])

# taking out K's and $ signs
minus_KnDollars = salary.apply(lambda x: x.replace('K','').replace('$',''))

#taking out 'per hour' and 'employer provided salary' texts from the salary estimate column
min_hr = minus_KnDollars.apply(lambda x: x.lower().replace('per hour', '').replace('employer provided salary:', ''))

# making another column for integer min salary
df['min_salary'] = min_hr.apply(lambda x: int(x.split('-')[0]))

# making another column for integer max salary
df['max_salary'] = min_hr.apply(lambda x: int(x.split('-')[1]))

# calculating average salary in a bew column
df['average_salary'] =(df.min_salary+df.max_salary)/2
```

```python
#COMPANY NAME TEXTS ONLY

#removed ratings from the company name
df['Company_text'] = df.apply(lambda x: x['Company Name'] if x['Rating']<0 else x['Company Name'][:-3], axis=1)

# a new column for state in which the job is made
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
df.job_state.value_counts()
```

```python
# a column that shows 1 if the job is at the company headquarters
df['Job_at_HQ'] = df.apply(lambda x: 1 if x.Location==x.Headquarters else 0, axis =1)

#age of the company
df['company_age'] = df['Founded'].apply(lambda x: 2020-x if x>0 else x)

```

```python
# JOB DESCRIPTION PARSING

#checking if python is mentioned in job description
df['python'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df.python.value_counts()

#checking if R is mentioned in job description
df['R_language'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() else 0)
df.R_language.value_counts()

#checking if spark is mentioned in job description
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df.spark.value_counts()

#checking if aws is mentioned in job description
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df.aws.value_counts()

#checking if excel is mentioned in job description
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df.excel.value_counts()
df.columns
```
